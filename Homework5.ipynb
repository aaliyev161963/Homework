# Case Intro
Term deposits are a major source of income for a bank. A term deposit is a cash investment held at a financial institution. Your money is invested for an agreed rate of interest over a fixed amount of time, or term. The bank has various outreach plans to sell term deposits to their customers such as email marketing, advertisements, telephonic marketing, and digital marketing.

Telephonic marketing campaigns still remain one of the most effective way to reach out to people. However, they require huge investment as large call centers are hired to actually execute these campaigns. Hence, it is crucial to identify the customers most likely to convert beforehand so that they can be specifically targeted via call.

The data is related to direct marketing campaigns (phone calls) of a Portuguese banking institution. The classification goal is to predict if the client will subscribe to a term deposit (variable y).

Content
The data is related to the direct marketing campaigns of a Portuguese banking institution. The marketing campaigns were based on phone calls. Often, more than one contact to the same client was required, in order to access if the product (bank term deposit) would be ('yes') or not ('no') subscribed by the customer or not. The data folder contains two datasets:-

Bank.csv: 45,211 rows and 18 columns ordered by date (from May 2008 to November 2010)

Detailed Column Descriptions
bank client data:

1 - age (numeric)

2 - job : type of job (categorical: "admin.","unknown","unemployed","management","housemaid","entrepreneur","student",
"blue-collar","self-employed","retired","technician","services")

3 - marital : marital status (categorical: "married","divorced","single"; note: "divorced" means divorced or widowed)

4 - education (categorical: "unknown","secondary","primary","tertiary")

5 - default: has credit in default? (binary: "yes","no")

6 - balance: average yearly balance, in euros (numeric)

7 - housing: has housing loan? (binary: "yes","no")

8 - loan: has personal loan? (binary: "yes","no")
# related with the last contact of the current campaign:
9 - contact: contact communication type (categorical: "unknown","telephone","cellular")
10 - day: last contact day of the month (numeric)

11 - month: last contact month of year (categorical: "jan", "feb", "mar", …, "nov", "dec")

12 - duration: last contact duration, in seconds (numeric)

# other attributes:
13 - campaign: number of contacts performed during this campaign and for this client (numeric, includes last contact)

14 - pdays: number of days that passed by after the client was last contacted from a previous campaign (numeric, -1 means client was not previously contacted)

15 - previous: number of contacts performed before this campaign and for this client (numeric)

16 - poutcome: outcome of the previous marketing campaign (categorical: "unknown","other","failure","success")

Output variable (desired target):

17 - y - has the client subscribed a term deposit? (binary: "yes","no")

Missing Attribute Values: None

```python
# Following function converts non-numeric variables (e.g., 'category', 'object') into numeric using label encoding
# Note:
#Label encoding converts categorical values into integer codes. Each unique category is assigned a distinct number, such as 1, 2, 3, 4, etc.
#Example : In our data, education variable has following values:{ 'primary' 'secondary', 'tertiary', 'unknown'}
#and it is mapped { 'primary': 0, 'secondary': 1, 'tertiary': 2, 'unknown': 3}

#  education  education_encoded
#  secondary                  1
#   tertiary                  2
#    primary                  0
#    unknown                  3
#   tertiary                  2
#    primary                  0
#
#Label encoding is generally suitable for tree-based models (e.g., decision trees, random forests, boosting methods).
#However, it may not be appropriate for models where the objective function relies on distance-based calculations, such as neural networks, support vector machines (SVM), or linear regression.
#For nominal features (categories with no intrinsic ordering, e.g., "red," "blue," "green"), label encoding can mislead the model by implying an ordinal relationship where none exists.
#In such cases, one-hot encoding is usually preferred.
#One-hot encoding creates binary columns (dummy variables) for each category in a categorical feature, avoiding the introduction of unintended ordinality.

def Encoder(df):
          from sklearn import preprocessing
          columnsToEncode = list(df.select_dtypes(include=['category','object']))
          le = preprocessing.LabelEncoder()
          for feature in columnsToEncode:
              try:
                  df[feature] = le.fit_transform(df[feature])
              except:
                  print('Error encoding '+feature)
          return df
```

```python
df=Encoder(df)
df
```

```python
y = df['y'] #Output
X = df.drop('y',axis=1)
X
```

```python
from sklearn.model_selection import  train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=17)
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score

# Initialize classifiers with default parameters and random_state for reproducibility
rf_classifier = RandomForestClassifier(random_state=17)
xgb_classifier = XGBClassifier(random_state=17, use_label_encoder=False, eval_metric='logloss') # suppress warning
lgbm_classifier = LGBMClassifier(random_state=17)
gb_classifier = GradientBoostingClassifier(random_state=17)

classifiers = {
    "Random Forest": rf_classifier,
    "XGBoost": xgb_classifier,
    "Light GBM": lgbm_classifier,
    "Gradient Boosting": gb_classifier
}

accuracy_results = {}

for name, classifier in classifiers.items():
    print(f"Training {name}...")
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_results[name] = accuracy
    print(f"{name} Accuracy on Test Data: {accuracy:.4f}\n")

# Find the best method
best_method = max(accuracy_results, key=accuracy_results.get)
best_accuracy = accuracy_results[best_method]

print(f"The method with the best accuracy on test data is {best_method} with an accuracy of {best_accuracy:.4f}")
```

```python
import optuna
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score

def objective_rf(trial):
    # Hyperparameters for RandomForestClassifier
    max_depth = trial.suggest_int("max_depth", 2, X_train.shape[1])
    max_features = trial.suggest_int("max_features", 2, X_train.shape[1])

    classifier = RandomForestClassifier(
        max_depth=max_depth,
        max_features=max_features,
        random_state=17
    )

    scores = cross_val_score(classifier, X_train, y_train, cv=3, scoring='accuracy')
    return scores.mean()

def objective_xgb(trial):
    # Hyperparameters for XGBClassifier
    max_depth = trial.suggest_int("max_depth", 2, X_train.shape[1])
    learning_rate = trial.suggest_float("learning_rate", 0.001, 0.3, log=True)

    classifier = XGBClassifier(
        max_depth=max_depth,
        learning_rate=learning_rate,
        random_state=17,
        use_label_encoder=False,
        eval_metric='logloss'
    )

    scores = cross_val_score(classifier, X_train, y_train, cv=3, scoring='accuracy')
    return scores.mean()

def objective_lgbm(trial):
    # Hyperparameters for LGBMClassifier
    max_depth = trial.suggest_int("max_depth", 2, X_train.shape[1])
    learning_rate = trial.suggest_float("learning_rate", 0.001, 0.3, log=True)

    classifier = LGBMClassifier(
        max_depth=max_depth,
        learning_rate=learning_rate,
        random_state=17
    )

    scores = cross_val_score(classifier, X_train, y_train, cv=3, scoring='accuracy')
    return scores.mean()

def objective_gb(trial):
    # Hyperparameters for GradientBoostingClassifier
    max_depth = trial.suggest_int("max_depth", 2, X_train.shape[1])
    learning_rate = trial.suggest_float("learning_rate", 0.001, 0.3, log=True)

    classifier = GradientBoostingClassifier(
        max_depth=max_depth,
        learning_rate=learning_rate,
        random_state=17
    )

    scores = cross_val_score(classifier, X_train, y_train, cv=3, scoring='accuracy')
    return scores.mean()


optuna_results = {}

print("Running Optuna study for Random Forest...")
study_rf = optuna.create_study(direction="maximize")
study_rf.optimize(objective_rf, n_trials=100)
optuna_results["Random Forest"] = {"best_params": study_rf.best_params, "best_cv_accuracy": study_rf.best_value}
print(f"Best parameters for Random Forest: {study_rf.best_params}")
print(f"Best CV accuracy for Random Forest: {study_rf.best_value:.4f}\n")

print("Running Optuna study for XGBoost...")
study_xgb = optuna.create_study(direction="maximize")
study_xgb.optimize(objective_xgb, n_trials=100)
optuna_results["XGBoost"] = {"best_params": study_xgb.best_params, "best_cv_accuracy": study_xgb.best_value}
print(f"Best parameters for XGBoost: {study_xgb.best_params}")
print(f"Best CV accuracy for XGBoost: {study_xgb.best_value:.4f}\n")

print("Running Optuna study for Light GBM...")
study_lgbm = optuna.create_study(direction="maximize")
study_lgbm.optimize(objective_lgbm, n_trials=100)
optuna_results["Light GBM"] = {"best_params": study_lgbm.best_params, "best_cv_accuracy": study_lgbm.best_value}
print(f"Best parameters for Light GBM: {study_lgbm.best_params}")
print(f"Best CV accuracy for Light GBM: {study_lgbm.best_value:.4f}\n")

print("Running Optuna study for Gradient Boosting...")
study_gb = optuna.create_study(direction="maximize")
study_gb.optimize(objective_gb, n_trials=100)
optuna_results["Gradient Boosting"] = {"best_params": study_gb.best_params, "best_cv_accuracy": study_gb.best_value}
print(f"Best parameters for Gradient Boosting: {study_gb.best_params}")
print(f"Best CV accuracy for Gradient Boosting: {study_gb.best_value:.4f}\n")


# Part a) Find best methods with parameters and best CV accuracy
best_overall_method = max(optuna_results, key=lambda k: optuna_results[k]['best_cv_accuracy'])
best_overall_params = optuna_results[best_overall_method]['best_params']
best_overall_cv_accuracy = optuna_results[best_overall_method]['best_cv_accuracy']

print(f"Q2 a) The method with the highest cross-validation accuracy is {best_overall_method}")
print(f"with best parameters: {best_overall_params}")
print(f"and best cross-validation accuracy: {best_overall_cv_accuracy:.4f}\n")

# Part b) Evaluate the performance on test data
if best_overall_method == "Random Forest":
    best_classifier = RandomForestClassifier(**best_overall_params, random_state=17)
elif best_overall_method == "XGBoost":
    best_classifier = XGBClassifier(**best_overall_params, random_state=17, use_label_encoder=False, eval_metric='logloss')
elif best_overall_method == "Light GBM":
    best_classifier = LGBMClassifier(**best_overall_params, random_state=17)
elif best_overall_method == "Gradient Boosting":
    best_classifier = GradientBoostingClassifier(**best_overall_params, random_state=17)

best_classifier.fit(X_train, y_train)
y_pred_optimized = best_classifier.predict(X_test)
accuracy_optimized = accuracy_score(y_test, y_pred_optimized)

print(f"Q2 b) Performance of {best_overall_method} with optimized parameters on test data:")
print(f"Accuracy: {accuracy_optimized:.4f}")

# Compare with default parameters from Q1
default_accuracy = accuracy_results[best_overall_method]
print(f"Default accuracy for {best_overall_method}: {default_accuracy:.4f}")

if accuracy_optimized > default_accuracy:
    print(f"Improvement: Yes, accuracy increased by {(accuracy_optimized - default_accuracy):.4f}")
elif accuracy_optimized < default_accuracy:
    print(f"Improvement: No, accuracy decreased by {(default_accuracy - accuracy_optimized):.4f}")
else:
    print("Improvement: No change in accuracy.")
```

Q3)Using Linear Regression,Decison Tree Random Forest,XGBoost, Light GBM and Gradient Boosting Classifier with default parameters (no parameter specifications except random_state) calculate R2 statistics on test data. Which method gives the best accuracy on test data

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.metrics import r2_score

# Initialize regressors with default parameters and random_state for reproducibility
linear_regressor = LinearRegression()
dt_regressor = DecisionTreeRegressor(random_state=17)
rf_regressor = RandomForestRegressor(random_state=17)
xgb_regressor = XGBRegressor(random_state=17, use_label_encoder=False, eval_metric='rmse') # suppress warning and use appropriate eval_metric
lgbm_regressor = LGBMRegressor(random_state=17)
gb_regressor = GradientBoostingRegressor(random_state=17)

regressors = {
    "Linear Regression": linear_regressor,
    "Decision Tree": dt_regressor,
    "Random Forest": rf_regressor,
    "XGBoost": xgb_regressor,
    "Light GBM": lgbm_regressor,
    "Gradient Boosting": gb_regressor
}

r2_results = {}

print("Q3) Calculating R2 statistics on Test Data:\n")

for name, regressor in regressors.items():
    print(f"Training {name}...")
    regressor.fit(X_train, y_train)
    y_pred = regressor.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    r2_results[name] = r2
    print(f"{name} R2 on Test Data: {r2:.4f}\n")

# Find the best method based on R2
best_r2_method = max(r2_results, key=r2_results.get)
best_r2_score = r2_results[best_r2_method]

print(f"The method with the best R2 on test data is {best_r2_method} with an R2 score of {best_r2_score:.4f}")
```

```python
import optuna
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.metrics import r2_score

def objective_rf_regressor(trial):
    # Hyperparameters for RandomForestRegressor
    max_depth = trial.suggest_int("max_depth", 2, X_train.shape[1])
    max_features = trial.suggest_int("max_features", 2, X_train.shape[1])

    regressor = RandomForestRegressor(
        max_depth=max_depth,
        max_features=max_features,
        random_state=17
    )

    scores = cross_val_score(regressor, X_train, y_train, cv=3, scoring='r2')
    return scores.mean()

def objective_xgb_regressor(trial):
    # Hyperparameters for XGBRegressor
    max_depth = trial.suggest_int("max_depth", 2, X_train.shape[1])
    learning_rate = trial.suggest_float("learning_rate", 0.001, 0.3, log=True)

    regressor = XGBRegressor(
        max_depth=max_depth,
        learning_rate=learning_rate,
        random_state=17,
        use_label_encoder=False, # Suppress warning
        eval_metric='rmse' # Appropriate eval_metric for regression
    )

    scores = cross_val_score(regressor, X_train, y_train, cv=3, scoring='r2')
    return scores.mean()

def objective_lgbm_regressor(trial):
    # Hyperparameters for LGBMRegressor
    max_depth = trial.suggest_int("max_depth", 2, X_train.shape[1])
    learning_rate = trial.suggest_float("learning_rate", 0.001, 0.3, log=True)

    regressor = LGBMRegressor(
        max_depth=max_depth,
        learning_rate=learning_rate,
        random_state=17
    )

    scores = cross_val_score(regressor, X_train, y_train, cv=3, scoring='r2')
    return scores.mean()

def objective_gb_regressor(trial):
    # Hyperparameters for GradientBoostingRegressor
    max_depth = trial.suggest_int("max_depth", 2, X_train.shape[1])
    learning_rate = trial.suggest_float("learning_rate", 0.001, 0.3, log=True)

    regressor = GradientBoostingRegressor(
        max_depth=max_depth,
        learning_rate=learning_rate,
        random_state=17
    )

    scores = cross_val_score(regressor, X_train, y_train, cv=3, scoring='r2')
    return scores.mean()
```

```python
optuna_reg_results = {}

print("Running Optuna study for Random Forest Regressor...")
study_rf_reg = optuna.create_study(direction="maximize")
study_rf_reg.optimize(objective_rf_regressor, n_trials=100)
optuna_reg_results["Random Forest Regressor"] = {"best_params": study_rf_reg.best_params, "best_cv_r2": study_rf_reg.best_value}
print(f"Best parameters for Random Forest Regressor: {study_rf_reg.best_params}")
print(f"Best CV R2 for Random Forest Regressor: {study_rf_reg.best_value:.4f}\n")

print("Running Optuna study for XGBoost Regressor...")
study_xgb_reg = optuna.create_study(direction="maximize")
study_xgb_reg.optimize(objective_xgb_regressor, n_trials=100)
optuna_reg_results["XGBoost Regressor"] = {"best_params": study_xgb_reg.best_params, "best_cv_r2": study_xgb_reg.best_value}
print(f"Best parameters for XGBoost Regressor: {study_xgb_reg.best_params}")
print(f"Best CV R2 for XGBoost Regressor: {study_xgb_reg.best_value:.4f}\n")

print("Running Optuna study for Light GBM Regressor...")
study_lgbm_reg = optuna.create_study(direction="maximize")
study_lgbm_reg.optimize(objective_lgbm_regressor, n_trials=100)
optuna_reg_results["Light GBM Regressor"] = {"best_params": study_lgbm_reg.best_params, "best_cv_r2": study_lgbm_reg.best_value}
print(f"Best parameters for Light GBM Regressor: {study_lgbm_reg.best_params}")
print(f"Best CV R2 for Light GBM Regressor: {study_lgbm_reg.best_value:.4f}\n")

print("Running Optuna study for Gradient Boosting Regressor...")
study_gb_reg = optuna.create_study(direction="maximize")
study_gb_reg.optimize(objective_gb_regressor, n_trials=100)
optuna_reg_results["Gradient Boosting Regressor"] = {"best_params": study_gb_reg.best_params, "best_cv_r2": study_gb_reg.best_value}
print(f"Best parameters for Gradient Boosting Regressor: {study_gb_reg.best_params}")
print(f"Best CV R2 for Gradient Boosting Regressor: {study_gb_reg.best_value:.4f}\n")


# Part a) Find best methods with parameters and best CV R2
best_overall_reg_method = max(optuna_reg_results, key=lambda k: optuna_reg_results[k]['best_cv_r2'])
best_overall_reg_params = optuna_reg_results[best_overall_reg_method]['best_params']
best_overall_reg_cv_r2 = optuna_reg_results[best_overall_reg_method]['best_cv_r2']

print(f"Q4 a) The method with the highest cross-validation R2 is {best_overall_reg_method}")
print(f"with best parameters: {best_overall_reg_params}")
print(f"and best cross-validation R2: {best_overall_reg_cv_r2:.4f}\n")

# Part b) Evaluate the performance on test data
if best_overall_reg_method == "Random Forest Regressor":
    best_regressor = RandomForestRegressor(**best_overall_reg_params, random_state=17)
elif best_overall_reg_method == "XGBoost Regressor":
    best_regressor = XGBRegressor(**best_overall_reg_params, random_state=17, use_label_encoder=False, eval_metric='rmse')
elif best_overall_reg_method == "Light GBM Regressor":
    best_regressor = LGBMRegressor(**best_overall_reg_params, random_state=17)
elif best_overall_reg_method == "Gradient Boosting Regressor":
    best_regressor = GradientBoostingRegressor(**best_overall_reg_params, random_state=17)

best_regressor.fit(X_train, y_train)
y_pred_optimized_reg = best_regressor.predict(X_test)
r2_optimized_reg = r2_score(y_test, y_pred_optimized_reg)

print(f"Q4 b) Performance of {best_overall_reg_method} with optimized parameters on test data:")
print(f"R2: {r2_optimized_reg:.4f}")

# Compare with default parameters from Q3
default_r2 = r2_results[best_overall_reg_method]
print(f"Default R2 for {best_overall_reg_method}: {default_r2:.4f}")

if r2_optimized_reg > default_r2:
    print(f"Improvement: Yes, R2 increased by {(r2_optimized_reg - default_r2):.4f}")
elif r2_optimized_reg < default_r2:
    print(f"Improvement: No, R2 decreased by {(default_r2 - r2_optimized_reg):.4f}")
else:
    print("Improvement: No change in R2.")
```
